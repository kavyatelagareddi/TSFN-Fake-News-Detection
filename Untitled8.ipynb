```python
# =========================
# STEP 1 — INSTALL
# =========================
!pip install sentence-transformers networkx

# =========================
# STEP 2 — EXTRACT ZIP
# =========================
import zipfile
import os

zip_path = "/content/archive (1).zip"   # your uploaded file
extract_path = "/content/FakeNewsNet"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction done!")

# =========================
# STEP 3 — LOAD JSON DATA
# =========================
import json

base_path = extract_path + "/politifact"   # you can also use gossipcop

news_data = []

for label_type in ["fake", "real"]:
    label = 1 if label_type == "fake" else 0
    folder_path = os.path.join(base_path, label_type)

    for news_id in os.listdir(folder_path):
        news_path = os.path.join(folder_path, news_id)

        content_file = os.path.join(news_path, "news content.json")
        if not os.path.exists(content_file):
            continue

        try:
            with open(content_file, "r") as f:
                data = json.load(f)
        except:
            continue

        text = data.get("text", "")

        tweets_file = os.path.join(news_path, "tweets.json")
        users = []

        if os.path.exists(tweets_file):
            try:
                with open(tweets_file, "r") as f:
                    tweets = json.load(f)

                for t in tweets:
                    if "user" in t and "id" in t["user"]:
                        users.append(str(t["user"]["id"]))
            except:
                pass

        if len(text) > 20:  # filter small text
            news_data.append({
                "text": text,
                "label": label,
                "users": users
            })

print("Loaded news:", len(news_data))


# =========================
# STEP 4 — SBERT EMBEDDINGS
# =========================
from sentence_transformers import SentenceTransformer
import numpy as np

print("Generating SBERT embeddings...")

sbert = SentenceTransformer('all-MiniLM-L6-v2')

texts = [n["text"] for n in news_data]
embeddings = sbert.encode(texts, show_progress_bar=True)


# =========================
# STEP 5 — BUILD USER TSFN (SIMPLIFIED)
# =========================
# assign random initial trust scores (simulate TSFN base)
user_tsfn = {}

for n in news_data:
    for u in n["users"]:
        if u not in user_tsfn:
            user_tsfn[u] = np.random.rand()


# =========================
# STEP 6 — GRAPH FEATURES
# =========================
import networkx as nx

G = nx.Graph()

for n in news_data:
    users = n["users"]
    for i in range(len(users) - 1):
        G.add_edge(users[i], users[i+1])

degree = dict(G.degree())


# =========================
# STEP 7 — BUILD FINAL FEATURES
# =========================
X_final = []
y_final = []

for i, n in enumerate(news_data):
    users = n["users"]

    if len(users) == 0:
        continue

    tsfn_vals = [user_tsfn[u] for u in users if u in user_tsfn]

    if len(tsfn_vals) == 0:
        continue

    mean_tsfn = np.mean(tsfn_vals)
    var_tsfn = np.var(tsfn_vals)

    weights = np.linspace(1, 0.5, len(tsfn_vals))
    weighted_mean = np.average(tsfn_vals, weights=weights)

    min_tsfn = np.min(tsfn_vals)
    max_tsfn = np.max(tsfn_vals)

    range_tsfn = max_tsfn - min_tsfn
    std_tsfn = np.std(tsfn_vals)

    # graph feature
    avg_degree = np.mean([degree.get(u, 0) for u in users])

    # combine SBERT + TSFN + graph
    features = list(embeddings[i]) + [
        mean_tsfn,
        var_tsfn,
        weighted_mean,
        min_tsfn,
        max_tsfn,
        range_tsfn,
        std_tsfn,
        avg_degree
    ]

    X_final.append(features)
    y_final.append(n["label"])

X_final = np.array(X_final)
y_final = np.array(y_final)

print("Final dataset shape:", X_final.shape)


# =========================
# STEP 8 — TRAIN MODEL
# =========================
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42
)

clf = XGBClassifier(n_estimators=100, max_depth=4)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))


# =========================
# STEP 9 — TEST NEW NEWS
# =========================
sample_text = ["Breaking news: something shocking happened"]

emb = sbert.encode(sample_text)

# dummy TSFN values
dummy = [0.5, 0.2, 0.8]

mean_tsfn = np.mean(dummy)
var_tsfn = np.var(dummy)

weights = np.linspace(1, 0.5, len(dummy))
weighted_mean = np.average(dummy, weights=weights)

min_tsfn = np.min(dummy)
max_tsfn = np.max(dummy)

range_tsfn = max_tsfn - min_tsfn
std_tsfn = np.std(dummy)

avg_degree = 1

test_features = list(emb[0]) + [
    mean_tsfn,
    var_tsfn,
    weighted_mean,
    min_tsfn,
    max_tsfn,
    range_tsfn,
    std_tsfn,
    avg_degree
]

prediction = clf.predict([test_features])

print("Final Prediction:", prediction)
```

GENERATE USERS+TWEETS

```python

def generate_users(text):
    num_users = random.randint(5, 15)
    return [f"user_{hash(text)%10000}_{i}" for i in range(num_users)]

def generate_tweets(text, users):
    return [text[:50] + f" tweet_by_{u}" for u in users]
```

STEP 5 — BUILD USER-TWEET DATA (STAGE 1 INPUT)

```python
news_data = []
all_tweets = []
tweet_users = []

for _, row in df.iterrows():
    text = str(row.get("text", row.get("title", "")))

    if text.strip() == "":
        continue

    users = generate_users(text)
    tweets = generate_tweets(text, users)

    news_data.append({
        "text": text,
        "label": row["label"],
        "users": users
    })

    all_tweets.extend(tweets)
    tweet_users.extend(users)

print("Total tweets:", len(all_tweets))
```

SBERT ON USER TWEETS

```python
from sentence_transformers import SentenceTransformer

sbert = SentenceTransformer('all-MiniLM-L6-v2')

tweet_embeddings = sbert.encode(all_tweets, show_progress_bar=True)
```

GRAPH USER NETWROK

```python
G = nx.Graph()

num_edges = len(tweet_users) * 2

for _ in range(num_edges):
    u = random.choice(tweet_users)
    v = random.choice(tweet_users)

    if u != v:
        G.add_edge(u, v)

degree = dict(G.degree())
```

```python
user_tsfn = {}

for user, score in zip(tweet_users, tsfn_final):
    user_tsfn[user] = score
```

```python
X = []
y = []

for n in news_data:

    users = n["users"][:5]  # First-K users

    if len(users) == 0:
        continue

    tsfn_vals = [user_tsfn.get(u, 0.5) for u in users]

    mean_tsfn = np.mean(tsfn_vals)
    var_tsfn = np.var(tsfn_vals)

    weights = np.linspace(1, 0.5, len(tsfn_vals))
    weighted_mean = np.average(tsfn_vals, weights=weights)

    min_tsfn = np.min(tsfn_vals)
    max_tsfn = np.max(tsfn_vals)

    range_tsfn = max_tsfn - min_tsfn
    std_tsfn = np.std(tsfn_vals)

    avg_degree = np.mean([degree.get(u, 0) for u in users])

    burstiness = var_tsfn / (mean_tsfn + 1e-5)

    features = [
        mean_tsfn,
        var_tsfn,
        weighted_mean,
        min_tsfn,
        max_tsfn,
        range_tsfn,
        std_tsfn,
        avg_degree,
        burstiness
    ]

    X.append(features)
    y.append(n["label"])

X = np.array(X)
y = np.array(y)

print("Final dataset:", X.shape)
```
